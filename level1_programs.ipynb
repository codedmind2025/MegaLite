```python
#Q13: Factorial (Iterative & Recursive)
num = 4
fact =1 
for i in range(1,num):
    fact = fact + fact *i
print(fact)

def Factorial(n):
    if n == 0 or n ==1:
        return 1
    else:
        return n*Factorial(n-1)
print(Factorial(4))
```

```python
#Q15: Fibonacci (Recursive)
def fibonnaci(num):
    if num == 0:
        return 0
    elif num ==1:
        return 1
    else:
        return fibonnaci(num-1) + fibonnaci(num-2)
    


print(fibonnaci(6))
```

```python
#Q17: LCM of Two Numbers
import math
a = 8
b = 2

# Product of inputs
ab = abs(a * b)

#gcd of inputs
greatest = math.gcd(a,b)

#lcm 
lcm_value = ab // greatest

print(lcm_value)
```

```python
#Q18: Check Armstrong Number
num = 153
temp = num
n = len(str(num))
fact = 0
while temp !=0:
    digits = temp % 10 
    fact = fact + digits**n
    temp = temp // 10
# print(fact)

# Check Armstrong
if(num == fact):
    print("Armstrong Numberr")
else:
    print("Not Armstrong Number")
```

```python
#Q19: Check Perfect Number
num = 6
sum_divisors = 0

for i in range(1, (num // 2) + 1):
    if num % i == 0:
        sum_divisors = sum_divisors + i

if sum_divisors == num:
    print(f"{num} is a Perfect Number")
else:
    print(f"{num} is Not a Perfect Number")

```
