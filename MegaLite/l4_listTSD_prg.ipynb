```python
#Q2: What looks like a copy , but isn't 
first = [1,2,3,4,5]
second = first
print(first)
second.append(6)
print(second)
print(first)

 # We use copy function
second = first.copy()
print(first)
second.append(8)
print(second)
print(first)
```

```python
#Q5: WAP to compute the average of a list of numbers
total = 0
count = 0
while(True):
    inp = input(f"Count Number {count} : ")
    if inp=="done": break
    value = float(inp)
    total = total + value
    count = count + 1
average = total /count
print(f"Average :",average)
```
