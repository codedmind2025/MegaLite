```python
#Q21: Calculate sum of all digits of a number.

num = int(input("Enter Number : "))
sum_digits = 0
while(num!=0):
    rem = num %10
    sum_digits += rem
    num//=10
print(sum_digits)
```

```python
#Q22: Compute base^exp without using built-in pow.
def Power(base,expo):
    if expo == 0:
        return 1
    else:
        return base*Power(base,expo-1)
# Input
base = int(input("Enter Base : "))
expo = int(input("Enter Exponent : "))

print("Result :",Power(base,expo))
```

```python
#Q23: Check if sum of factorials of digits equals the number.
num = 145
temp = num
sum_fact = 0
while(temp >0):
    digit = temp %10

    #factorial
    fact = 1
    for i in range(1,digit+1):
        fact = fact *i
        # print(i,fact)
    sum_fact +=fact
    # print(sum_fact)
    temp //=10
# print(sum_fact)


if(sum_fact == num):
    print("Strong number")
else:
    print("Weak Number ")
```
