```python
class Parent:
    def Greet(self, name):
        self.name = name
        return self.name

class Child(Parent):
    def Print(self):
        result = f"hello, {self.name}"
        return result

# Create an instance of Child
obj = Child()

# Call the Greet method from Parent
name = input("Enter Name : ")
obj.Greet(name)

# Call the Print method from Child
print(obj.Print())
```
