```python
#Q13: Compute n! both iteratively and recursively.

def Factorial(num):
    if num == 0 or num == 1:
        return 1
    else:
        return num*Factorial(num-1)
Factorial(5)
```

```python
#Q15: Find the Nth Fibonacci number using recursion.
def Fibonacci(num):
    if num == 1:
        return 0
    elif num == 2:
        return 1
    else:
        return Fibonacci(num-1)+Fibonacci(num-2)
num = 10
for i in range(1,num+1):
    print(Fibonacci(i))
```

```python
#Q17: Find LCM using the GCD relation.
import math
def gcd_num(a,b):
    #inputs ko mulitply krna
    lcm_values =abs(a*b)
    #gcd get krna hai
    gcd_values = math.gcd(a,b)
    lcm= lcm_values//gcd_values
    return lcm
gcd_num(6,12)
```

```python
#Q18: Check if a number equals sum of cubes of its digits.
num = 153
original_num=  num
Sum_digits = 0
n = len(str(num))
for i in range(n):
    temp = num % 10
    Sum_digits += temp**n
    num = num//10

print(Sum_digits)
# Check the number is Armstrong
if(original_num== Sum_digits):
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```
