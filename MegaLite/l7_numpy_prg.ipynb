### Creation of Arrays with Evenly spaced values 

```python
# Arange  function use 
# Syntax of arange
# => arange([start,] stop[,step], [,dtype=None])


import numpy as np
a = np.arange(1,10)
print(a)
x = range(1,10)
print(x)
print(list(x))
# further arange examples
x = np.arange(12)
print(x)
x = np.arange(1,20,2,float) # start : Stop : step : datatype
print(x)

```

```python
# Linspace
# The Syntax of linspace 
#=> linspace(start,stop,num=50,endpoint =True,retstep=False)

# 50 values between 1 and 10
print(np.linspace(1,10)) # Default it gives you 50 values between 1 & 10 
# 7 values betweeen 1 and 10
print(np.linspace(1,10,7)) # you can set manual the how much require

print(len(np.linspace(1,10,7))) # Length of the values

# excluding the endpoint
print(np.linspace(1,10,7,endpoint=False))
```

```python
# Zero Dimensional Arrays in Numpy 
x = np.array(42)
print("x: ",x)
print("The type of x : ",type(x))
print("The dimension of x : ",np.ndim(x))
```
