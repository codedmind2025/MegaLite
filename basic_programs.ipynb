```python
#Q6 Perform +, -, *, / based on user input.
num1 = int(input("Enter Num1: "))
num2 = int(input("Enter Num2: "))
print(f"="*10 , "Simple Calculator" , "="*10)
add = num1 + num2
sub = num1 - num2
mul = num1 * num2

print(f"Addition is {add}")
print(f"Subtraction is {sub}")
print(f"Multiplication is {mul}")

if num2 == 0:
    print("Division is not possible (ZeroDivisionError)")
else:
    div = num1 / num2
    print(f"Division is {div}")
```

```python
#Q8 Compute area of circle, rectangle, and triangle.
pi = 3.17
radius = 4
length = 20
breadth = 30
side = 10
print("1. Area of Circle ")
print("2. Area of Rectangle")
print("3. Area of Triangle")

choice = int(input("> Choose from above :"))

# condition
if(choice == 1):
    area_circle = pi * radius**2
    print(area_circle)
elif(choice == 2):
    area_rectangle = length * breadth
    print(area_rectangle)
elif(choice == 3):
    area_triangle = side **2
    print(area_triangle)
elif(choice == ""):
    print("Please Enter Something")
else:
    print("Enter Valid Choice !")
    
```

```python
#Q9 Count how many digits a number has.
num = 134
count = 0
while(num > 0):
    count = count + 1
    num //=10
print(count)
```

```python
#Q10 Reverse digits of an integer (e.g., 1234 → 4321).
num = 1342
rev = 0
while(num > 0):
    rem = num % 10 
    rev = rev*10 + rem
    num=num//10

print(rev)
```
