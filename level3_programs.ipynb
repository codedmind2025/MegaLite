```python
## Pattern Printing ##
#Q26: Right Triangle Star Pattern
num = 5
for i in range(1,num+1):
    for j in range(1,i+1):
        print("*", end=" ")
    print( )
```

```python
#Q27: Inverted Triangle Pattern
num = 5
for i in range(0,num+1):
    for j in range(1,num-i+1):
        print("*",end=" ")
    print( )
```

```python
#Q28: Right angle traingle 
num = 5
for i in range(0,num):
    for j in range(0,num-i):
        print(" ",end=" ")
    for k in range(0,i+1):
        print("*",end=" ")
    print()
```

```python
#Q29: Pyramid pattern
num=5
for i in range(1,num+1):
    for j in range(num-i):
        print(" ",end="")
    for k in range(2*i-1):
        print("*",end="")
    print()
```

```python
#Q31: Print a diamond shape using stars.
num = 3
for i in range(1,num+1):
    for j in range(i-1):
        print("",end=" ")
    for k in range(i-1,num):
        print("*",end=" ")
    print(" ")
```
